## Plotting static visualizations
Now we import matplotlib to plot and our main class. Then, we obtain a static visualization of the policy and Infected class.

```python
import matplotlib.pyplot as plt
import pyvisvsm.pyvisvsm as vsm


fig = plt.figure()
vs_fig = vsm.VsmFigures(fig, data_dir="../data/")
vs_fig.load_data()
ax_states, fig_states = vs_fig.plot_states()
ax_policy, fig_policy = vs_fig.plot_policies()
fig, ax1, ax2, ax3, ax4 = vs_fig.dash_plot()
plt.show()
```

## Building a dynamic visualization: Animating and recording

To deploy the animation, we need to write and configure the parameters using the class method `VSMAnimator.rec`.

```python
vs_an = vsm.VsmAnimator(fig)
print(vs_an.data_folder)
vs_an.set_up_axes(data_dir='../data/')
vs_an.load_data()
vs_an.record_animation()
```
